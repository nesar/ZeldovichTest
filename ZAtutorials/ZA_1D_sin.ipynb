```python
import numpy as np
import matplotlib.pylab as plt
```

```python
q = 2*np.pi*np.linspace(-1.0,2.0,200)   # Lagrangian - grid coords
n = 2./3 

def D(t, n): return np.float(t)**(n) # Linear growth factor
def s(q): return np.sin(q) # s(q) =  -grad_q[phi(q)] 
def x(q, t, n): return q - D(t, n)*s(q)  # Euler coords
def rho(q, t, n): return (1-D(t, n)*np.cos(q))**(-1)  # density perturbation
def delta(q,t,n): return -D(t,n)*np.gradient(s(q))
```

```python
def neglog(a): #return symlog(a)
    neg = np.ones_like(a)
    for i in range(np.size(a)):
        if a[i]==0:  neg[i] = 0 
        if a[i]>0: neg[i] = np.log10(a[i])
        if a[i]<0: neg[i] = -np.log10(-a[i])
    return neg
```

```python
for time in [0.1 ,1.000000001, 2.0]:

    plt.figure(2, figsize=(18, 12))
    plt.subplot(311)
    plt.plot(q, x(q, time, n), label = "t = %d" %time, lw=2)  # q vs x
    #plt.xlabel('Lagrangian coordinate q')
    plt.ylabel(r'$x$')
    plt.xlim(0, 10)
    
    plt.figure(2)
    plt.subplot(312)
    plt.plot(q, n*D(time, n-1)*s(q), label = "t = %d" %time, lw=2)  # S vs q
    #plt.xlabel('Lagrangian coordinate q')
    plt.ylabel(r'$v$')
    plt.xlim(0, 10)
    
    
    plt.figure(3,  figsize=(18, 12))
    plt.subplot(211)
    plt.plot(x(q, time, n), n*D(time, n-1)*s(q), label = "t = %d" %time, lw=2)  # S vs q
    #plt.xlabel('Euler coordinate x')
    plt.ylabel(r'$v$')
    plt.xlim(0, 10)

    
    plt.figure(2)
    plt.subplot(313)

    plt.plot( q , (np.abs(rho(q, 0.01+time, n)/rho(q, 0.01, n)-1)) ,  label = "t = %d" %time)  # rho vs x
    plt.yscale('log')
    
    plt.xlabel(r'$q$')
    plt.ylabel(r'$\rho/\rho_0$')
    plt.xlim(-1, 10)
    
    plt.figure(3)
    plt.subplot(212)
    plt.plot(x(q, 0.01+time, n) , (np.abs(rho(q, 0.01+time, n)/rho(q, 0.01, n)-1)) ,  label = "t = %d" %time)  # rho vs x
    plt.yscale('log')
    
    plt.xlabel(r'$x$')
    plt.ylabel(r'$\rho/\rho_0$')
    plt.xlim(-1, 10)

############################################

plt.figure(2)
plt.subplot(3,1,2)
plt.axvline( x = np.pi, linewidth=1, color='k', ls = '--')


plt.figure(3)
plt.subplot(2,1,2)
plt.axvline( x = np.pi, linewidth=1, color='k', ls = '--')


plt.figure(2)
plt.subplot(3,1,1)
plt.axvline( x = np.pi, linewidth=1, color='k', ls = '--')


plt.figure(3)
plt.subplot(2,1,1)
plt.axvline( x = np.pi, linewidth=1, color='k', ls = '--')


plt.figure(2)
plt.subplot(3,1,3)
plt.axvline( x = np.pi, linewidth=1, color='k', ls = '--')
plt.legend(loc = 'upper center', ncol = 3)


plt.figure(3)
plt.subplot(2,1,2)
plt.axvline( x = np.pi, linewidth=1, color='k', ls = '--')
plt.legend(loc = 'upper center', ncol = 3)



plt.show()
```
